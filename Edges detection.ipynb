```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

Menampilkan Citra

```python
image = cv.imread("/content/WhatsApp Image 2022-02-23 at 2.18.03 PM.jpeg")
plt.imshow(image)
```

Menerapkan detekksi tepi menggunakan operator prewit terhadap horizontal



```python
filter = [[-1,-2,-1], [0,0,0],[1,2,1]]
n,m,d = image.shape
matrix = np.zeros_like(image)

for row in range(3,n-2):
  for col in range(3,m-2):
    local = image[row-1:row+2, col-1:col+2,0]
    transform = filter*local
    vs = (transform.sum() + 4)/8
    matrix[row,col] = [vs]*3

plt.imshow(matrix)
```

Menerapkan detekksi tepi menggunakan operator prewit terhadap vertikal

```python
filter = [[1,1,1], [0,0,0],[-1,-1,-1]]
n,m,d = image.shape
matrix = np.zeros_like(image)

for row in range(3,n-2):
  for col in range(3,m-2):
    local = image[row-1:row+2, col-1:col+2,0]
    transform = filter*local
    vs = (transform.sum() + 4)/8
    matrix[row,col] = [vs]*3

plt.imshow(matrix)
```
