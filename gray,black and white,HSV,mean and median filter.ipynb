Tugas 2 Pengolahan Citra Digital

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread ("/content/16.jpeg")
```

```python
from google.colab.patches import cv2_imshow
cv2_imshow(img)
```

```python
gray_img = cv2.cvtColor (img, cv2.COLOR_BGR2GRAY)
cv2_imshow(gray_img)
```

```python
(thresh, im_bw) = cv2.threshold(gray_img, 128, 255, cv2.THRESH_BINARY | cv2.THRESH_OTSU)
cv2_imshow(im_bw)
```

```python
import cv2
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2_imshow(hsv_img)
```

Mean filter

```python
import numpy as np
from scipy.signal import convolve2d

# Mean kernel
ksize = 3
kernel = np.ones((ksize,ksize))
kernel /= kernel.sum()

im_out = convolve2d(gray_img, kernel)
cv2_imshow(im_out)
```

Gaussian filter

```python
from scipy.signal import convolve2d
from skimage.filters import gaussian

# Mean vs Gaussian
im_out2 = gaussian(gray_img, 3)
cv2_imshow(im_out2)
```

median filter

```python
import numpy as np
import matplotlib.pyplot as plt
# Median filter
ksize = 3
img_gray = gray_img
im_out3 = np.zeros(img_gray.shape)
for y in range(ksize//2, img_gray.shape[0]-ksize//2):
    for x in range(ksize//2, img_gray.shape[1]-ksize//2):
        im_out3[y,x] = np.median(img_gray[y-ksize//2:y+ksize//2+1, x-ksize//2: x+ksize//2+1])

plt.figure()
plt.gray()
plt.subplot(1,2,1)
plt.title('Mean filter')
plt.imshow(im_out)#menampilkan citra dengan mean filters menggunakan kernel matriks 3 kali 3
plt.subplot(1,2,2)
plt.title('Median filter')
plt.imshow(im_out3)#menampilkan citra dengan median filters menggunakan kernel matriks 3 kali 3
plt.show()
```
