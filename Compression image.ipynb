Mengimport library dan menampilkan citra

```python
from skimage import io
from sklearn.cluster import KMeans
import numpy as np
image = io.imread('/content/WhatsApp Image 2022-02-01 at 6.59.51 PM.jpeg')
io.imshow(image)
io.show()
```

```python
#Membuat dimensi pada citra
rows = image.shape[0]
cols = image.shape[1]

image = image.reshape(rows*cols, 3)

#Implementasikan kmeans clustering dengan 20 kluster
kmeans = KMeans(n_clusters=20)
kmeans.fit(image)

#Mengganti tiap pixel dengan centroid terdekat
compressed_image = kmeans.cluster_centers_[kmeans.labels_]
compressed_image = np.clip(compressed_image.astype('uint8'), 0, 255)

compressed_image = compressed_image.reshape(rows, cols, 3)

#Save and display output image
io.imsave('Terkompresi_citra.jpeg', compressed_image)
io.imshow(compressed_image)
io.show()
```

Ada beberapa metode yang digunakan dalam melakukan kompresi pada citra digital, salah satunya dengan menggunakan k-means clustering. Dengan menggunakan k means clustering banyaknya gambar pada citra direduksi menjadi beberapa kluster, tiap kluster memiliki RGB yang bernilai sama yang membuat ukuran penyimpanan dari citra tersebut mengecil.
